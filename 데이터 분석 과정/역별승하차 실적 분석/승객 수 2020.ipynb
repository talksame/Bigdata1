```python
역별승하차실적_2020 = pd.read_csv('역별 승하차실적_2020.csv')
subset = 역별승하차실적_2020[['역무열차종','정차역','하차인원수']]
subset['승객수'] = pd.to_numeric(subset['하차인원수'], errors='coerce').fillna(0)
sub20 =subset[['역무열차종','정차역','승객수']]
sub20=sub20.groupby(by=['정차역'] ,as_index=False).sum()
sub20=sub20.sort_values(by='승객수',ascending=False)
sub20=sub20.iloc[0:10]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 

sub20.plot(x="정차역", y=["승객수"], kind="bar",figsize=(10,9))
plt.ylabel('승객수(백만)')
plt.ylim([0, 2500000])
plt.xticks(rotation=0)
plt.title("2020년 정차역별 승객수")
plt.show()
```
