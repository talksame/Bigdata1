```python
data = pd.read_csv('/content/drive/MyDrive/관광거점도시/국내 지역별 문화체육관광시설 인근 쇼핑시설(백화점, 면세점, 할인매장 등) 데이터.csv')
```

```python
data
```

```python
data.fillna('None')
```

```python
def general_jj(a):
  if "전주" in a:
    return "전주시"
  else:
    return a
```

```python
def get_market(a):
  if "백화점" in a :
    return "대형백화점"
  elif "에브리데이" in a:
    return "지역슈퍼/마켓"
  elif "홈플러스익스프레스" in a or "GS슈퍼마켓" in a or "하나로마트" in a or "롯데슈퍼" in a:
    return "기업형마트" 
  elif "JOYMART" in a  or "농협하나로클럽" in a  or "OK마트" in a  or "ABC마트" in a :
    return "중형마트"
  elif "성)홈플러스" in a  or "이마트" in a  or "롯데마트" in a  or "하이마트" in a :
    return "대형마트"
  else:
    return a
```

```python
data['mcate_nm'] = data['mcate_nm'].apply(get_market)
data = data[["sgg_nm", "mcate_nm"]]
#a = data[data['sido_nm'].str.contains("전라") | data['sido_nm'].str.contains("강원") | data['sido_nm'].str.contains("경상") | data['sido_nm'].str.contains("충청")] 
a['sgg_nm'] = a['sgg_nm'].apply(general_jj)
general_data = a[a['sgg_nm'].str.contains("전주") | a['sgg_nm'].str.contains("강릉")|a['sgg_nm'].str.contains("목포")]
general_data=general_data[["sgg_nm", "mcate_nm"]]
```

```python
counting_shopping = general_data.groupby(["sgg_nm"]).count()
counting_shopping = counting_shopping.rename(columns={'mcate_nm':'쇼핑'})
counting_shopping
```

```python
data['mcate_nm'].unique()
```

```python
def get_cafe(a):
  if "투썸플레이스" in a or "크리스피크림도넛" in a or "공차" in a or "뚜레쥬르" in a or "탐앤탐스" in a or "설빙" in a or "배스킨라빈스" in a or "드롭탑" in a :
    return "카페(도심)"
  elif "기타아이스크림" in a:
    return "아이스크림"
  elif "스타벅스" in a :
    return "스타벅스"
  elif "나뚜루" in a or "로티보이" in a or "한스델리" in a or "할리스커피" in a or "카페띠아모" in a or "커핀그루나루" in a or "카페베네" in a or "이디야" in a :
    return "카페(도심)" 

  elif "신라명과" in a or "커피빈" in a or "카페쎄리오" in a or "파스쿠찌" in a or "엔제리너스" in a or "파리바게뜨" in a or "주커피" in a or "자바시티" in a :
    return "카페(도심)"    

  elif "스무디킹" in a or "망고식스" in a or "카페아모제" in a  or '요거프레소' in a or '던킨' in a:
    return "카페(도심)"    
  elif "차" in a:
    return "전통찻집"  

  else:
    return a
```

```python
data['mcate_nm'] = data['mcate_nm'].apply(get_cafe)
#data = data[["sido_nm", "sgg_nm", "mcate_nm"]]
#a = data[data['sido_nm'].str.contains("전라") | data['sido_nm'].str.contains("강원") | data['sido_nm'].str.contains("경상") ] 
a['sgg_nm'] = a['sgg_nm'].apply(general_jj)
general_data = a[a['sgg_nm'].str.contains("전주") | a['sgg_nm'].str.contains("강릉")|a['sgg_nm'].str.contains("목포")]
counting_cafe = general_data.groupby(["sgg_nm"]).count()
counting_cafe=counting_cafe.rename(columns={'sido_nm':'카페'})
counting_cafe
```

```python
import matplotlib.pyplot as plt
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.5, 'edgecolor': 'w', 'linewidth': 5}
plt.pie(counting_cafe.sido_nm, labels=counting_cafe.index	,autopct='%.1f%%',colors=colors, wedgeprops=wedgeprops, counterclock=False)
plt.title('카페', pad=-110) 
plt.show()
```
